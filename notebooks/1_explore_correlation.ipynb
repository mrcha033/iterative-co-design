# Exploring Activation Correlations

This notebook demonstrates how to compute an activation correlation matrix for a model layer. Replace the toy example with real model activations when available.

```

import torch
import matplotlib.pyplot as plt
import seaborn as sns

# Example: generate random activations for 8 dimensions
activations = torch.randn(100, 8)

# Compute Pearson correlation matrix
corr = torch.corrcoef(activations.T)

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Activation Correlation Matrix')
plt.show()
```
